```python
import pandas as pd
from nltk.stem.snowball import EnglishStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import Image

clean_df_all= pd.read_pickle("clean_df_all.pkl")
```

```python
'''We are using TF-IDF and Cosine Similarity, So if any synonyms that can be used by the user has to be
included in the dictionary so we can get good result. How to identify the synonyms, by going through user queries
and checking which among them are not getting the right result and including it in the synonym_dict '''

synonym_dict={
    'power up':['switch on'],
}
```

```python
def search_solution(query):
    flag_sol_zero=0
    search_query=query
    stemmed=' '.join([EnglishStemmer().stem(token) for token in search_query.split()])
    ''' For changing synonyns'''    
    for key,value in synonym_dict.items():
        for i in value:
            if stemmed.find(i)!=-1:
                stemmed=stemmed.replace(i,key)
    vec = TfidfVectorizer()
    doclist_tfidf = vec.fit_transform(clean_df_all['Problem']).toarray()
    query_tfidf = vec.transform([stemmed]).toarray()
    rec_idx = cosine_similarity(doclist_tfidf, query_tfidf)
    clean_df_all['cos_sim']=rec_idx
    #_all.sort_values(['cos_sim'],ascending=False)
    print('\n\nProblem: '+search_query+'\n\nSolution:')
    try:
        solution=clean_df_all[clean_df_all.cos_sim>0.05].sort_values(['cos_sim'],ascending=False).reset_index()['Solution'][0]
    except:
        flag_sol_zero=1
    if(flag_sol_zero==0):
        if(solution.find('SampleDocuments/pic_references/')!=-1):
            display(Image(solution))
        else:
            print(solution)
    else:
        print('Sorry. Please enter something meaningful.')
```

```python
queries=['Login Page is not getting displayed',
'Login Page Down',
'Login Page not opening',
'Login Page is not showing up.',
'Login credetials are not working.',
'Forgot the password, can you please reset the password for me.',
'Password expired.',
'FW: Myspace Login Issue',
'Please reset mypassword.It is expired.',
'Account got locked, because of 3 wrong attempts os password. Could you please reset the password',
'Account locked. Not able to login',
'Create new account and share the username and password',
'New Account is needed for a new joinee.',
'New Account Creation for user Tom.',
'Needs permission wifi access to laptop',
'Wifi access needed.',
'Need to connect to Wifi.',
'Unable to connect to Internet',
'How can I Delete Internet Cookies',
'Not able to send and receive emails',
'Computer is not getting switched on',
'Computer not powering up',
'Volume Icon is not showing in right corner']
```

```python
for i in queries:
    search_solution(i)
    print('-'*100)
```
